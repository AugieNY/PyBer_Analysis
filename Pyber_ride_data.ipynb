```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import statistics
```

```python
pyber_ride_df=pd.read_csv("Pyber_ride_data.csv")
pyber_ride_df
```

```python
pyber_ride_df.plot(x="Month",y="Avg. Fare ($USD)")
```

```python
# Set x-axis and tick locations.
x_axis = np.arange(len(pyber_ride_df))
tick_locations = [value for value in x_axis]

#Plot the data
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)")
plt.xticks(tick_locations, pyber_ride_df["Month"])
plt.show()
```

```python
pyber_ride_df.plot.bar(x="Month", y="Avg. Fare ($USD)")
```

```python
#Add error bars.
errorbar = pyber_ride_df["Avg. Fare ($USD)"].std()

pyber_ride_df.plot.bar(x="Month", y="Avg. Fare ($USD)",yerr=errorbar)
#Add caps to the error bars.
pyber_ride_df.plot.bar(x="Month", y="Avg. Fare ($USD)",yerr=errorbar,capsize = 3)

#Change the color of the bars to sky blue.
pyber_ride_df.plot.bar(x="Month", y="Avg. Fare ($USD)",yerr=errorbar,capsize = 3, color="skyblue")
#Rotate the labels on the x-axis to horizontal.

#Set the y-axis increment to every $5.
plt.yticks(np.arange(0,60,step=5))
```
